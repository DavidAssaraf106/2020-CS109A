```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Data generating for Lorentz model 
```

```python
def Lorentz(w,w0=1, sig=0.0):
    Gamma = 7e-2
    eps = 1 - w0**2/(w0**2-w**2+1j*Gamma) + sig*np.random.randn(w.shape[0])
    return  eps.real
```

```python
x = np.linspace(0,2,64*2)
y = Lorentz(x)
plt.plot(x,y,'o')
```

```python
data1={'x': x, 'y':y}
df = pd.DataFrame(data1,columns=['x','y'])
export_csv = df.to_csv('../data/lorentz_set.csv', index=None, header=True)
```

#### Noise data

```python
x = np.linspace(0,2,64*2)
y = Lorentz(x, sig=0.8)
plt.plot(x,y,'o')
```

```python
data1={'x': x, 'y':y}
df = pd.DataFrame(data1,columns=['x','y'])
export_csv = df.to_csv('../data/lorentz_noise_set.csv', index=None, header=True)
```

```python
def room1(x,sig=0.0):
    y = np.abs(x-.5)*x-.9*np.abs(x-.5)**3 + sig*np.random.randn(x.shape[0])
    return  y
```

```python
x = np.linspace(0,2,64*2)
y = room1(x, sig=0.1)
plt.plot(x,y,'o')
```

```python
data1={'x': x, 'y':y}
df = pd.DataFrame(data1,columns=['x','y'])
export_csv = df.to_csv('../data/breakRoom1.csv', index=None, header=True)
```
